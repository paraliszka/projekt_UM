```python
X = df.drop('Spending Score (Normalized)', axis=1)
y = df['Spending Score (Normalized)']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print('Błąd średniokwadratowy: ', mse)
print('Współczynnik determinacji (R-squared): ', r2)
print('Pierwiastek z błędu średniokwadratowego: ', rmse)
print('Średni błąd bezwzględny', mae)
```

```python
model.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.03),
    metrics=[
        tf.keras.metrics.BinaryAccuracy(name='accuracy'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall'),
    ],
)
```

```python
X_train_scaled = preprocessing_pipeline.fit_transform(X_train)
X_test_scaled = preprocessing_pipeline.transform(X_test)
```

```python
model.build((None, X_train_scaled.shape[1]))
```

```python
model.summary()
```

```python
plot_model(model, show_shapes=True)
```

```python
history = model.fit(X_train_scaled, y_train, epochs=50)
```

```python
rcParams['figure.figsize'] = (18, 8)

plt.plot(
    np.arange(1, 51),
    history.history['loss'], label='Loss'
)
plt.plot(
    np.arange(1, 51),
    history.history['accuracy'], label='Accuracy'
)
plt.plot(
    np.arange(1, 51),
    history.history['precision'], label='Precision'
)
plt.plot(
    np.arange(1, 51),
    history.history['recall'], label='Recall'
)
plt.title('Evaluation metrics', size=20)
plt.xlabel('Epoch', size=14)
plt.legend()
```
